**The map() Function in Python**

The map() function in Python is a built-in function used to apply a given function to each item of an iterable (like a list, tuple, or string) and return a new iterator with the results. It is particularly useful for performing element-wise transformations or computations on a collection of data.


**Answer-10**

The map(), reduce(), and filter() functions in Python are all functional programming tools that operate on iterables. Each serves a distinct purpose and is used for specific types of transformations or computations.

1. map()

Purpose:

 Applies a given function to all items in an iterable (like a list) and returns a new iterable (a map object) with the results.


Syntax: map(function, iterable)

2. reduce()

Purpose: Applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, reducing the iterable to a single value. It’s available in the functools module.


Syntax: reduce(function, iterable, [initial])

3. filter()

Purpose: Filters the items in an iterable based on a function that returns True or False. It returns a new iterable containing only the elements that evaluate to True.


Syntax: filter(function, iterable)

```python
def sum_of_even_numbers(numbers):

    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return sum(even_numbers)

numbers = [47, 11, 42, 13, 8, 22]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
#Answer-4

def is_prime(number):
    if number <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False  # If divisible by any number other than 1 and itself, it's not prime
    return True  # If no divisors are found, the number is prime


for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

```python
#Answer-5

class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms to generate
        self.a, self.b = 0, 1  # Starting values for the Fibonacci sequence
        self.count = 0  # Counter to keep track of how many terms have been generated

    def __iter__(self):
        return self  # The iterator itself is returned

    def __next__(self):
        if self.count < self.n:
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b  # Update the values for the next Fibonacci number
            self.count += 1
            return fib_number
        else:
            raise StopIteration  # Stop iteration when the specified number of terms is reached

# Example usage
n_terms = 10
fibonacci_sequence = FibonacciIterator(n_terms)

for num in fibonacci_sequence:
    print(num)
```

```python
#Answer-6

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i


exponent = 5
for power in powers_of_two(exponent):
    print(power)
```

```python
#Answer-10

# Function to check if a character is a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

input_string = "Hello, World!"

filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```
